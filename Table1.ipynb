```python
import pandas as pd
import statsmodels.api as sm
```

```python
climate_conflict = pd.read_csv("clim_conflict_for_R.csv")
climate_conflict
```

```python
icc_columns = [col for col in climate_conflict.columns if col.startswith("iccode")]
print(icc_columns)
```

```python
icc_year_columns = [col for col in climate_conflict.columns if col.startswith("Iccyear")]
print(icc_year_columns)
```

```python
#Sort Dataframe 
climate_conflict_table1 = climate_conflict[
    ["country"]+["year_actual"]+["war_prio_new"]+["temp_all"]+["temp_all_lag"]+
    ["prec_all"]+["prec_all_lag"]+["gdp_l"]+["polity2_lag"]+["year"]+["ccode"]+icc_columns+icc_year_columns
]
```

```python
#Define lists for the fixed effects variables in the table 1 regressions 
icc_columns_mo = [col for col in climate_conflict_table1.columns if col.startswith("iccode")]
icc_year_columns_mo = [col for col in climate_conflict_table1.columns if col.startswith("Iccyear")]
```

```python
climate_conflict_model1 = climate_conflict_table1.dropna(subset=["temp_all","temp_all_lag"])
climate_conflict_model2 = climate_conflict_table1.dropna(subset=["temp_all","temp_all_lag","prec_all","prec_all_lag"])
climate_conflict_model3 = climate_conflict_table1.dropna(subset=["temp_all","temp_all_lag","prec_all","prec_all_lag","gdp_l","polity2_lag","year"])
```

```python
#Pick variables for model 1
X = climate_conflict_model1[["temp_all"]+["temp_all_lag"]+icc_columns_mo+icc_year_columns_mo]
y = climate_conflict_model1["war_prio_new"]
#Add constant 
X = sm.add_constant(X)
#Create model 1 regression
model1 = sm.OLS(y,X).fit(cov_type='cluster', cov_kwds={'groups': climate_conflict_model1['ccode']})
#Summary
model1.summary()
```

```python
#Pick variables for model 2
X = climate_conflict_model2[["temp_all"]+["temp_all_lag"]+["prec_all"]+["prec_all_lag"]+icc_columns_mo+icc_year_columns_mo]
y = climate_conflict_model2["war_prio_new"]
#Add constant 
X = sm.add_constant(X)
#Create model 2 regression
model2 = sm.OLS(y,X).fit(cov_type='cluster', cov_kwds={'groups': climate_conflict_model2['ccode']})
#Summary
model2.summary()
```

```python
#Pick variables for model 3
X = climate_conflict_model3[["temp_all"]+["temp_all_lag"]+["prec_all"]+["prec_all_lag"]+
icc_columns_mo+["gdp_l"]+["polity2_lag"]+["year"]
]
y = climate_conflict_model3["war_prio_new"]
#Add constant 
X = sm.add_constant(X)
#Create model 3 regression
model3 = sm.OLS(y,X).fit(cov_type='cluster', cov_kwds={'groups': climate_conflict_model3['ccode']})
#Summary
model3.summary()
```
